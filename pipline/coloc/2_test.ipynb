```python
from finemap_tools.finemap.coloc import coloc_Wrapper
```

```python
topLoci = (1, 15583355, 15583355)  # (chr, start, end)
locus_range = 1500  # 100kb
locus_range = locus_range * 1000
locus_range_tuple = (topLoci[0], topLoci[1] - locus_range, topLoci[2] + locus_range)

locus_region = f"{locus_range_tuple[0]}:{locus_range_tuple[1]}-{locus_range_tuple[2]}"
print(f"locus_region: {locus_region}")
```

```python
gwas_file = "/home/xutingfeng/GIFT/data/GWAS/T1Mapping_Cortex_20240129.csv_firstorder_Median_all_2023_GRCh38_unionKidneys.tsv.gz"
save_dir = "/home/xutingfeng/GIFT/data/analysis/DNAJC16/coloc"
from pathlib import Path

Path(save_dir).mkdir(parents=True, exist_ok=True)
eQTL_path = "/home/xutingfeng/GIFT/GTEx_V8/ge/Kidney_Cortex.tsv.gz"
```

```python
from finemap_tools.others.polyfun.polyfun_utils import (
    configure_logger,
)

configure_logger("./log.txt")
```

```python
coloc = coloc_Wrapper(
    sumstats_file1=gwas_file,
    sumstats_file1_format="format",
    sumstats_file2=eQTL_path,
    sumstats_file2_format="GTExV8",
    n1=4786,
    n2=73,
    type1="quant",
    sdY1=1,
    chr_num=locus_range_tuple[0],
    start=locus_range_tuple[1],
    end=locus_range_tuple[2],
)
```

```python
gene_col = "gene_id"
# drop duplicated SNPs
sumstats1_duplicated_num = coloc.sumstats_file1.duplicated(subset=["snp"]).sum()
sumstats2_duplicated_num = coloc.sumstats_file2.duplicated(
    subset=["snp", gene_col]
).sum()
```

```python
# if sumstats1_duplicated_num > 0:
#     logging.warning(
#         f"sumstats1 have {sumstats1_duplicated_num} duplicated SNPs and will be dropped"
#     )
# if sumstats2_duplicated_num > 0:
#     logging.warning(
#         f"sumstats2 have {sumstats2_duplicated_num} duplicated SNPs and will be dropped"
#     )

coloc.sumstats_file1 = coloc.sumstats_file1.drop_duplicates(subset=["snp"])
coloc.sumstats_file2 = coloc.sumstats_file2.drop_duplicates(subset=["snp", gene_col])

assert (
    coloc.sumstats_file1.shape[0] > 0 and coloc.sumstats_file2.shape[0] > 0
), "no valid SNPs to run coloc after drop duplicated SNPs"

# sync sumstats
coloc.sync_sumstats()
```
